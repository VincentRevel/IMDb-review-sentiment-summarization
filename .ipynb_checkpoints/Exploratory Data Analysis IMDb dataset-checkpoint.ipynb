```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

from datasets import load_dataset
import torch
import time
import evaluate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from bs4 import BeautifulSoup
import re
```

```python
dataset = load_dataset("imdb")
```

```python
dataset
```

```python
# Function to clean text from HTML tags
def clean_text(text):
    # Remove HTML tags
    text = BeautifulSoup(text, 'html.parser').get_text()
    # Remove escape characters
    text = re.sub(r'\\', '', text)
    return text
```

```python
# Clean the text in the dataset
for split in ['train', 'test', 'unsupervised']:
    dataset[split] = dataset[split].map(lambda example: {'text': clean_text(example['text']), 'label': example['label']})
```

```python
print("First few rows of the dataset:")
for i in range(3):
    print(dataset['train']['text'][i])
    dash_line = '-'.join('' for x in range(100))
    print(dash_line)
```

```python
# Class distribution
labels = dataset['train']['label']
positive_count = sum(labels)
negative_count = len(labels) - positive_count
plt.figure(figsize=(6, 6))
plt.pie([positive_count, negative_count], labels=['Positive', 'Negative'], autopct='%1.1f%%')
plt.title('Class Distribution')
plt.show()
```

```python
# Sentiment analysis
positive_samples = [example['text'] for example, label in zip(dataset['train'], labels) if label == 1]
negative_samples = [example['text'] for example, label in zip(dataset['train'], labels) if label == 0]
print("Number of positive samples:", len(positive_samples))
print("Number of negative samples:", len(negative_samples))
```

```python
# Text length distribution
text_lengths = [len(example['text']) for example in dataset['train']]
plt.figure(figsize=(10, 6))
plt.hist(text_lengths, bins=50)
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.title('Text Length Distribution')
plt.show()
```

```python
# Word cloud
all_texts = ' '.join([example['text'] for example in dataset['train']])
wordcloud = WordCloud(width=800, height=400).generate(all_texts)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud')
plt.show()
```
